```python
import pandas   #importing module 
df = pandas.read_csv('/content/drive/MyDrive/mnit folder/Data1_Assignment1.xlsx - Sheet2 - Data1_Assignment1.xlsx - Sheet2.csv')  #reading csv file using pandas library
df.fillna(value = df.mean(), inplace = True)    #replacing "Nan" value with mean values using "df.mean" 
```

```python
gtype =  int(input("Choose the digit for type of category: 1-Fasting, 2-before meal, 3-After 1 to 2 hours of food intake, 4-at bedtime"))

l = len(df['Blood Glucose (mg/dL)'])

#g =['']*l
d =['']*l
y1 =  []
y2 =[]
c1 = []
c2 = []
c3 =[]
g = df['Blood Glucose (mg/dL)']
```

```python
if gtype ==1:
  for i in range(0,24):
    if g[i] in range(0,101):
      d[i] = 'not Diabetic'
      c1.append(i)
      y1.append(g[i])
    elif g[i] == 'Not available':
      c3.append(i)
    else :
      d[i] = 'Diabetic'
      c2.append(i)
      y2.append(g[i])

elif gtype ==2:
  for i in range(0,24):
    if g[i] in range(70,131):
      d[i] = 'not Diabetic'
      c1.append(i)
      y1.append(g[i])
    elif g[i] == 'Not available':
      c3.append(i)
    else :
      d[i] = 'Diabetic'
      c2.append(i)
      y2.append(g[i])

elif gtype ==3:
  for i in range(0,24):
    if g[i] in range(0,181) :
      d[i] = 'not Diabetic'
      y1.append(g[i])
      c1.append(i)
    elif g[i] == 'Not available':
      c3.append(i)
    else :
      d[i] = 'Diabetic'
      c2.append(i)
      y2.append(g[i])

elif gtype ==4:
  for i in range(0,24):
    if g[i] in range(100,141):
      d[i] = 'not Diabetic'
      c1.append(i)
      y1.append(g[i])
    elif g[i] == 'Not available':
      c3.append(i)
    else :
      d[i] = 'Diabetic'
      c2.append(i)
      y2.append(g[i])
```

```python
import matplotlib.pyplot as plt
plt.bar(c1, y1, label="Normal", color='b')
plt.bar(c2, y2, label="Diabetic", color='r')
plt.plot()

plt.xlabel("serial no")
plt.ylabel("Glucose level")
plt.title("Bar Chart Example")
plt.legend()
plt.show()
```
